```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w1 = 1/(2*np.pi)
w2 = 1
w3 = 2*np.pi
x = np.linspace(0, 10, 500)
```

```python
# Define the functions to plot

def f1(x):
    return 1 + np.exp(-2*x) - 2*np.exp(-x)*np.cos(w1*x)

def f2(x):
    return 1 + np.exp(-2*x) - 2*np.exp(-x)*np.cos(w2*x)

def f3(x):
    return 1 + np.exp(-2*x) - 2*np.exp(-x)*np.cos(w3*x)

```

```python
plt.xlabel('Time')
plt.ylabel('Probability')
plt.title('Probability for the first excited state  for t>=0 ')

plt.plot(x, f1(x), label='w = 1/2pi')
plt.plot(x, f2(x), label='w = 1')
plt.plot(x, f3(x), label='w = 2pi')

plt.legend()

plt.show()
```
